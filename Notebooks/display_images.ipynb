```
def display_images(my_images):
    row, col = 12, 6
    my_images = my_images[:row * col]
    my_images = np.reshape(my_images, (my_images.shape))
    _, ax = plt.subplots(row, col, figsize=(col * 4, row * 4))
    for i, image in enumerate(my_images):
        ax[i // col, i % col].imshow(image, vmin=0, vmax=1, cmap="bone")
        ax[i // col, i % col].axis("off")
    plt.show()
    plt.close()
```

```
display_images(images)
```

```
display_images(labels)
```

```
display_images(preds)
```
