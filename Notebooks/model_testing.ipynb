```
import datetime
import os
import numpy as np
import tensorflow as tf
```

```
np.random.seed = 1

dataset = "DRIVE"
dim = 96 if dataset == "DRIVE" else 192

main_path = f"/content/gdrive/My Drive"
```

```
testing_images = f"{main_path}/Data/{dataset}/patches_{dim}/images/test/"

test_img = next(os.walk(testing_images))[2]
test_img.sort()

X_test = np.concatenate([np.load(testing_images + file_id)["arr_0"] for file_id in test_img], axis=0)

X_test = X_test / 255
```

```
X_test.shape
```

```
model = tf.keras.models.load_model(f"{main_path}/NN Models/{dataset}/model_DRIVE_EffnetB7_R2_96_f")
block = f"R2"
```

```
# model.summary()
```

```
preds = model.predict(X_test)
```

```
preds.shape
```

```
np.savez_compressed(f"{main_path}/Data/{dataset}/patches_{dim}/predictions/preds_EffnetB7_{block}_{str(datetime.datetime.now())[:-7]}.npz", preds)
```
