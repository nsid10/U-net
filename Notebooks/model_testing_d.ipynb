```
import datetime
import os
import numpy as np
import tensorflow as tf
```

```
np.random.seed = 42

main_path = f"/content/gdrive/My Drive"

dataset = "ISIC 2018"

sub_path = f"{main_path}/Data/{dataset}/256/"
```

```
X_test = np.load(f"{sub_path}test_images.npz")["arr_0"]

X_test = X_test / 255
```

```
print(np.max(X_test))
print(np.min(X_test))
```

```
X_test.shape
```

```
model = tf.keras.models.load_model(f"{main_path}/NN Models/{dataset}/checkpoint_final_R2-deep5")
block = f"R2_deep5"
```

```
model.summary()
```

```
preds = model.predict(X_test)
```

```
preds.shape
```

```
np.savez_compressed(f"{main_path}/Data/{dataset}/256/predictions/preds_EffnetB7_{block}_{str(datetime.datetime.now())[:-7]}.npz", preds)
```
